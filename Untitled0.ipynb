```python
from rdkit import Chem
from rdkit.Chem import Draw

smiles = {
    "DMT": "CN(C)CCc1c[nH]c2ccccc12",
    "Harmina": "COc1ccc2c(c1)[nH]c3c2nccc3",
    "Harmalina": "COc1ccc2c(c1)N=C3C=CCN3C2",
    "Triptamina": "NCCc1c[nH]c2ccccc12"
}

mols = [Chem.MolFromSmiles(s) for s in smiles.values()]
Draw.MolsToGridImage(mols, legends=list(smiles.keys()), molsPerRow=2)
```
